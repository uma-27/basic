```python
#preprossing
#1.resize
#2.flattening
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize


target=[]
images=[]
flat_data=[]


DATADIR='/content/images'
CATEGORIES=['dry soil','wet soil']

for category in CATEGORIES:
  print(category)
  class_num=CATEGORIES.index(category)
  path = os.path.join(DATADIR,category)
  for img in os.listdir(path):
    img_array=imread(os.path.join(path,img))
    print(img_array)
    print(img_array.shape)
    plt.imshow(img_array)
    img_resized=resize(img_array,(150,150,3)) #normalizes 0 to 1
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)

flat_data=np.array(flat_data)
target=np.array(target)
images=np.array(images)
```

```python
np.unique(target,return_counts=True)
unique,count=np.unique(target,return_counts=True)
plt.bar(CATEGORIES,count)
```

```python
#split data into framing and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flat_data,target,test_size=0.3,random_state=100)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid=[
             {'C':[1,10,100,1000],'kernel':['linear']},
             {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']},
]

svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
#save the model using pickel library
import pickle
pickle.dump(clf,open('img_model.p','wb'))
```
